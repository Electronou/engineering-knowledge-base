## 1️⃣ Variables & Data Types

### 📝 Notes
- A **variable** is a named container that stores data.
- Python uses **dynamic typing** — you don't declare types explicitly.
- Variable names are **case-sensitive** (`age` ≠ `Age`).

### Core Data Types
| Type | Example | Description |
|------|---------|-------------|
| `int` | `42` | Whole numbers |
| `float` | `3.14` | Decimal numbers |
| `str` | `"hello"` | Text |
| `bool` | `True/False` | Boolean |
| `NoneType` | `None` | Absence of value |

### Naming Rules
- Start with a letter or underscore `_`
- Cannot start with a number
- No spaces (use `snake_case`)
- Cannot use Python keywords (`if`, `for`, `while`, etc.)

````python
# ── Integer ──
age = 25
year = 2024

# ── Float ──
pi = 3.14159
temperature = -4.5

# ── String ──
name = "Alice"
city = 'Hyderabad'

# ── Boolean ──
is_student = True
has_passed = False

# ── NoneType ──
result = None

# Check types with type()
print(type(age))         # <class 'int'>
print(type(pi))          # <class 'float'>
print(type(name))        # <class 'str'>
print(type(is_student))  # <class 'bool'>
print(type(result))      # <class 'NoneType'>
````

````python
# ── Type Conversion (Casting) ──
x = "42"
print(int(x))        # str → int  → 42
print(float(x))      # str → float → 42.0

y = 3.99
print(int(y))        # float → int → 3 (truncates, not rounds!)

z = 100
print(str(z))        # int → str  → '100'
print(bool(0))       # 0 is False
print(bool(42))      # Non-zero is True
print(bool(""))      # Empty string is False
print(bool("hi"))    # Non-empty string is True
````

````python
# ── Multiple Assignment ──
a, b, c = 1, 2, 3
print(a, b, c)   # 1 2 3

# Same value to multiple variables
x = y = z = 0
print(x, y, z)   # 0 0 0

# Swap values (Pythonic!)
a, b = 10, 20
a, b = b, a
print(a, b)      # 20 10
````

## 2️⃣ Operators

### 📝 Notes
Operators perform operations on values/variables.

| Category | Operators |
|----------|-----------|
| Arithmetic | `+  -  *  /  //  %  **` |
| Comparison | `==  !=  >  <  >=  <=` |
| Logical | `and  or  not` |
| Assignment | `=  +=  -=  *=  /=  //=  **=  %=` |
| Identity | `is  is not` |
| Membership | `in  not in` |
| Bitwise | `&  |  ^  ~  <<  >>` |

````python
# ── Arithmetic Operators ──
a, b = 17, 5

print(a + b)    # Addition      → 22
print(a - b)    # Subtraction   → 12
print(a * b)    # Multiplication→ 85
print(a / b)    # Division      → 3.4  (always float)
print(a // b)   # Floor Div     → 3    (removes decimal)
print(a % b)    # Modulus       → 2    (remainder)
print(a ** b)   # Exponent      → 1419857 (17^5)
````

````python
# ── Comparison Operators (return bool) ──
x, y = 10, 20

print(x == y)   # Equal             → False
print(x != y)   # Not Equal         → True
print(x > y)    # Greater than      → False
print(x < y)    # Less than         → True
print(x >= 10)  # Greater or Equal  → True
print(x <= 5)   # Less or Equal     → False
````

````python
# ── Logical Operators ──
a, b = True, False

print(a and b)   # Both must be True → False
print(a or b)    # At least one True → True
print(not a)     # Negation          → False

# Real-world example
age = 22
has_id = True
can_enter = age >= 18 and has_id
print("Can enter:", can_enter)   # True
````

````python
# ── Identity Operators ──
a = [1, 2, 3]
b = a          # b points to same object
c = [1, 2, 3]  # c is a different object

print(a is b)      # True  (same memory)
print(a is c)      # False (different object)
print(a == c)      # True  (same value)

# ── Membership Operators ──
fruits = ["apple", "banana", "cherry"]
print("apple" in fruits)       # True
print("mango" not in fruits)   # True
````

## 3️⃣ Strings

### 📝 Notes
- Strings are **immutable** sequences of characters.
- Created with single `'`, double `"`, or triple `'''` / `"""` quotes.
- **Indexing** starts at `0` from left, `-1` from right.
- **Slicing** syntax: `string[start:stop:step]`

````python
# ── String Creation ──
s1 = 'Hello'
s2 = "World"
s3 = '''This is a
multi-line string'''

# ── Concatenation & Repetition ──
greeting = s1 + ", " + s2 + "!"
print(greeting)          # Hello, World!
print("Ha" * 3)          # HaHaHa

# ── Indexing & Slicing ──
word = "Python"
#        P y t h o n
#        0 1 2 3 4 5   (forward)
#       -6-5-4-3-2-1   (backward)

print(word[0])       # P
print(word[-1])      # n
print(word[0:3])     # Pyt   (index 0,1,2)
print(word[::2])     # Pto   (every 2nd char)
print(word[::-1])    # nohtyP (reversed!)
````

````python
# ── Common String Methods ──
s = "  Hello, Python World!  "

print(s.strip())         # Remove leading/trailing whitespace
print(s.lower())         # Lowercase all
print(s.upper())         # Uppercase all
print(s.title())         # Title Case

sentence = "apple,banana,cherry"
print(sentence.split(","))   # ['apple', 'banana', 'cherry']
print("-".join(["a","b","c"]))  # a-b-c

text = "I love Python"
print(text.replace("Python", "coding"))  # I love coding
print(text.find("love"))     # 2  (index where found)
print(text.count("o"))       # 2
print(text.startswith("I"))  # True
print(text.endswith("on"))   # True
print(len(text))             # 13
````

````python
# ── String Formatting ──
name = "Alice"
score = 95.678

# Method 1: f-strings (Python 3.6+) ← RECOMMENDED
print(f"Name: {name}, Score: {score:.2f}")

# Method 2: .format()
print("Name: {}, Score: {:.2f}".format(name, score))

# Method 3: % operator (old style)
print("Name: %s, Score: %.2f" % (name, score))

# f-string expressions
a, b = 10, 3
print(f"{a} + {b} = {a+b}")     # 10 + 3 = 13
print(f"{name!r}")               # 'Alice' (repr)
print(f"{score:10.2f}")          # Right-aligned in 10 chars
````

## 4️⃣ Lists

### 📝 Notes
- **Ordered**, **mutable** (changeable), allows **duplicates**.
- Can hold items of **mixed types**.
- Defined with square brackets `[]`.
- Zero-indexed like strings.

````python
# ── Creating Lists ──
empty = []
numbers = [1, 2, 3, 4, 5]
mixed   = [42, "hello", 3.14, True, None]
nested  = [[1, 2], [3, 4], [5, 6]]

# ── Indexing & Slicing ──
fruits = ["apple", "banana", "cherry", "date", "elderberry"]
print(fruits[0])       # apple
print(fruits[-1])      # elderberry
print(fruits[1:4])     # ['banana', 'cherry', 'date']
print(fruits[::-1])    # reversed list

# ── Modifying Lists ──
fruits[1] = "blueberry"   # Update
print(fruits)
````

````python
# ── List Methods ──
nums = [3, 1, 4, 1, 5, 9, 2, 6]

nums.append(7)         # Add to end
nums.insert(0, 0)      # Insert at index 0
nums.remove(1)         # Remove FIRST occurrence of 1
popped = nums.pop()    # Remove & return last item
popped2 = nums.pop(2)  # Remove & return item at index 2

print(nums)
print("Popped:", popped, popped2)

nums.sort()            # Sort in-place (ascending)
print("Sorted:", nums)

nums.sort(reverse=True)  # Descending
print("Desc:", nums)

nums.reverse()         # Reverse in-place
print("Reversed:", nums)

print("Length:", len(nums))
print("Count of 1:", nums.count(1))
print("Index of 9:", nums.index(9))
````

````python
# ── List Operations ──
a = [1, 2, 3]
b = [4, 5, 6]

# Concatenation
print(a + b)          # [1, 2, 3, 4, 5, 6]

# Extend
a.extend(b)
print(a)              # [1, 2, 3, 4, 5, 6]

# Copy (shallow)
original = [1, 2, 3]
copy1 = original.copy()
copy2 = original[:]    # Another way

# Nested list access
matrix = [[1, 2, 3], [4, 5, 6], [7, 8, 9]]
print(matrix[1][2])    # 6  (row 1, col 2)
````

## 5️⃣ Tuples

### 📝 Notes
- **Ordered**, **immutable** (cannot change after creation), allows **duplicates**.
- Defined with parentheses `()` or just commas.
- Faster than lists; use when data should not change.
- Great for **unpacking** and **returning multiple values** from functions.

````python
# ── Creating Tuples ──
empty_tuple = ()
single = (42,)          # MUST have trailing comma for 1-element tuple!
coords = (10.5, 20.3)
rgb    = (255, 128, 0)
mixed  = (1, "two", 3.0, True)

# Packing (implicit tuple)
point = 3, 4            # same as (3, 4)

# ── Unpacking ──
x, y = coords
print(f"x={x}, y={y}")

r, g, b = rgb
print(f"Red={r}, Green={g}, Blue={b}")

# Extended unpacking
first, *rest = (1, 2, 3, 4, 5)
print(first)   # 1
print(rest)    # [2, 3, 4, 5]

*beginning, last = (1, 2, 3, 4, 5)
print(beginning)  # [1, 2, 3, 4]
print(last)       # 5
````

````python
# ── Tuple Methods ──
t = (1, 2, 3, 2, 4, 2)

print(t.count(2))   # 3  (count occurrences)
print(t.index(3))   # 2  (first index of value)
print(len(t))       # 6
print(min(t))       # 1
print(max(t))       # 4
print(sum(t))       # 14

# Tuple is immutable — this will error:
try:
    t[0] = 99
except TypeError as e:
    print(f"Error: {e}")

# But you can convert tuple ↔ list
lst = list(t)
lst[0] = 99
t2 = tuple(lst)
print(t2)
````

## 6️⃣ Sets

### 📝 Notes
- **Unordered**, **mutable**, **no duplicates**.
- Defined with curly braces `{}` or `set()`.
- Very fast for **membership testing** (`in` operator).
- Supports mathematical set operations.

````python
# ── Creating Sets ──
empty_set = set()         # NOT {} — that creates a dict!
fruits = {"apple", "banana", "cherry"}
nums = {1, 2, 3, 4, 5}

# Duplicates are auto-removed
s = {1, 2, 2, 3, 3, 3}
print(s)    # {1, 2, 3}

# ── Set Methods ──
s.add(4)           # Add element
s.discard(2)       # Remove (no error if not present)
s.remove(1)        # Remove (raises KeyError if not present)
print(s)

# Membership
print(3 in s)      # True
print(99 in s)     # False
````

````python
# ── Set Operations ──
A = {1, 2, 3, 4, 5}
B = {4, 5, 6, 7, 8}

print("Union:",        A | B)          # All unique elements
print("Intersection:", A & B)          # Elements in BOTH
print("Difference:",   A - B)          # In A but NOT in B
print("Sym Diff:",     A ^ B)          # In A or B but NOT both

# Subset / Superset
C = {1, 2, 3}
print(C.issubset(A))    # True  (all C in A)
print(A.issuperset(C))  # True  (A contains all C)

# Use case: Remove duplicates from list
lst = [1, 2, 2, 3, 4, 4, 5]
unique = list(set(lst))
print(unique)
````

## 7️⃣ Dictionaries

### 📝 Notes
- **Ordered** (Python 3.7+), **mutable**, **key-value pairs**, **no duplicate keys**.
- Keys must be **immutable** (string, int, tuple).
- Defined with `{}` using `key: value` syntax.
- Extremely fast for lookups by key — O(1) average.

````python
# ── Creating Dictionaries ──
empty = {}
student = {
    "name": "Alice",
    "age": 21,
    "grade": "A",
    "scores": [90, 85, 92]
}

# Using dict() constructor
config = dict(host="localhost", port=5432, debug=True)

# ── Accessing Values ──
print(student["name"])              # Direct access (KeyError if missing)
print(student.get("age"))           # Safe access → 21
print(student.get("email", "N/A"))  # Default if key missing → N/A

# ── Add / Update / Delete ──
student["email"] = "alice@example.com"   # Add new key
student["age"] = 22                       # Update existing key
del student["grade"]                      # Delete key
popped = student.pop("email")             # Remove & return value
print(student)
````

````python
# ── Dictionary Methods ──
info = {"name": "Bob", "age": 30, "city": "Mumbai"}

print(info.keys())    # dict_keys(['name', 'age', 'city'])
print(info.values())  # dict_values(['Bob', 30, 'Mumbai'])
print(info.items())   # dict_items([('name','Bob'), ...])

# Iterating
for key, value in info.items():
    print(f"{key}: {value}")

# Merging dicts (Python 3.9+)
d1 = {"a": 1, "b": 2}
d2 = {"c": 3, "d": 4}
merged = d1 | d2
print(merged)

# update() method
d1.update(d2)
print(d1)
````

````python
# ── Dictionary Comprehension ──
squares = {x: x**2 for x in range(1, 6)}
print(squares)   # {1: 1, 2: 4, 3: 9, 4: 16, 5: 25}

# Filter: only even squares
even_sq = {x: x**2 for x in range(1, 11) if x % 2 == 0}
print(even_sq)

# Nested dict
employees = {
    "emp1": {"name": "Alice", "dept": "Engineering"},
    "emp2": {"name": "Bob",   "dept": "Design"}
}
print(employees["emp1"]["name"])   # Alice
````

## 8️⃣ Conditional Statements

### 📝 Notes
- Python uses `if`, `elif`, `else` for branching logic.
- **Indentation** (4 spaces) defines code blocks — no `{}` needed!
- **Ternary operator**: `value_if_true if condition else value_if_false`
- **match-case** (Python 3.10+) is the `switch` equivalent.

````python
# ── Basic if / elif / else ──
score = 75

if score >= 90:
    grade = "A"
elif score >= 80:
    grade = "B"
elif score >= 70:
    grade = "C"
elif score >= 60:
    grade = "D"
else:
    grade = "F"

print(f"Score: {score} → Grade: {grade}")

# ── Ternary (One-liner) ──
age = 20
status = "Adult" if age >= 18 else "Minor"
print(status)

# ── Nested if ──
x = 15
if x > 0:
    if x % 2 == 0:
        print(f"{x} is positive and even")
    else:
        print(f"{x} is positive and odd")
else:
    print(f"{x} is non-positive")
````

````python
# ── Truthy & Falsy Values ──
# Falsy: 0, 0.0, '', [], {}, (), None, False
# Truthy: Everything else

values = [0, 1, "", "hello", [], [1], None, True]
for v in values:
    if v:
        print(f"{v!r:10} → Truthy")
    else:
        print(f"{v!r:10} → Falsy")
````

````python
# ── match-case (Python 3.10+) ── 
import sys

if sys.version_info >= (3, 10):
    command = "quit"
    match command:
        case "start":
            print("Starting...")
        case "stop" | "quit":
            print("Stopping/Quitting")
        case _:
            print("Unknown command")
else:
    print("match-case requires Python 3.10+")
````

## 9️⃣ Loops

### 📝 Notes
- **`for` loop**: iterate over sequences (list, string, range, dict, etc.)
- **`while` loop**: repeat as long as condition is `True`
- **`break`**: exit the loop immediately
- **`continue`**: skip current iteration, go to next
- **`pass`**: placeholder — do nothing
- **`else` on loops**: runs if loop completes normally (no `break`)

````python
# ── for loop with range() ──
for i in range(5):         # 0, 1, 2, 3, 4
    print(i, end=" ")
print()

for i in range(1, 10, 2):  # 1, 3, 5, 7, 9
    print(i, end=" ")
print()

for i in range(10, 0, -2): # 10, 8, 6, 4, 2
    print(i, end=" ")
print()

# ── Iterating over sequences ──
fruits = ["apple", "banana", "cherry"]
for fruit in fruits:
    print(f"🍎 {fruit}")

# ── enumerate() — get index + value ──
for idx, fruit in enumerate(fruits, start=1):
    print(f"{idx}. {fruit}")
````

````python
# ── while loop ──
count = 0
while count < 5:
    print(count, end=" ")
    count += 1
print()

# ── break & continue ──
print("\nbreak example:")
for i in range(10):
    if i == 5:
        break          # Stop when i == 5
    print(i, end=" ")

print("\ncontinue example:")
for i in range(10):
    if i % 2 == 0:
        continue       # Skip even numbers
    print(i, end=" ")

# ── Loop else ──
print("\n\nloop-else:")
for i in range(5):
    if i == 10:        # Never true
        break
else:
    print("Loop completed without break!")  # This runs
````

````python
# ── zip() — iterate multiple sequences ──
names  = ["Alice", "Bob", "Charlie"]
scores = [92, 85, 78]

for name, score in zip(names, scores):
    print(f"{name}: {score}")

# ── Nested Loops ──
print("\nMultiplication table (1-3):")
for i in range(1, 4):
    for j in range(1, 4):
        print(f"{i}×{j}={i*j}", end="  ")
    print()
````

## 🔟 Functions

### 📝 Notes
- Defined with `def` keyword.
- **Parameters**: variables in function definition.
- **Arguments**: actual values passed when calling.
- Use `return` to send a value back. Without it, returns `None`.
- **Docstrings** `"""..."""` document what a function does.

````python
# ── Basic Function ──
def greet(name):
    """Return a greeting message."""
    return f"Hello, {name}!"

print(greet("Alice"))
help(greet)   # Shows docstring

# ── Default Parameters ──
def power(base, exponent=2):
    return base ** exponent

print(power(4))      # 4^2 = 16  (uses default)
print(power(2, 10))  # 2^10 = 1024

# ── Keyword Arguments ──
def describe(name, age, city):
    return f"{name}, {age} years old, from {city}"

print(describe(age=25, city="Delhi", name="Bob"))  # order doesn't matter
````

````python
# ── *args — Variable positional arguments ──
def add_all(*numbers):
    """Sum any amount of numbers."""
    return sum(numbers)

print(add_all(1, 2))           # 3
print(add_all(1, 2, 3, 4, 5)) # 15

# ── **kwargs — Variable keyword arguments ──
def profile(**details):
    """Print any key-value details."""
    for key, val in details.items():
        print(f"  {key}: {val}")

profile(name="Alice", age=25, hobby="coding")

# ── Combined ──
def full_example(pos, *args, keyword="default", **kwargs):
    print(f"pos={pos}, args={args}, keyword={keyword}, kwargs={kwargs}")

full_example(1, 2, 3, keyword="custom", x=10, y=20)
````

````python
# ── Scope: Local vs Global ──
x = "global"

def show_scope():
    x = "local"   # Different x, local to this function
    print("Inside function:", x)

show_scope()         # Inside function: local
print("Outside:", x) # Outside: global

# Using global keyword
counter = 0

def increment():
    global counter
    counter += 1

increment()
increment()
print("Counter:", counter)  # 2

# ── Return multiple values ──
def min_max(lst):
    return min(lst), max(lst)   # Returns a tuple!

lo, hi = min_max([3, 1, 7, 2, 9])
print(f"Min={lo}, Max={hi}")
````

## 1️⃣1️⃣ Lambda Functions

### 📝 Notes
- **Anonymous** (nameless) functions defined in one line.
- Syntax: `lambda arguments: expression`
- Can only have **one expression** (implicit return).
- Commonly used with `map()`, `filter()`, `sorted()`.

````python
# ── Basic Lambda ──
square = lambda x: x ** 2
add    = lambda a, b: a + b
greet  = lambda name: f"Hello, {name}!"

print(square(5))       # 25
print(add(3, 7))       # 10
print(greet("Alice"))  # Hello, Alice!

# ── map() — apply function to each element ──
numbers = [1, 2, 3, 4, 5]
squared = list(map(lambda x: x**2, numbers))
print("Squared:", squared)

# ── filter() — keep elements where function is True ──
evens = list(filter(lambda x: x % 2 == 0, numbers))
print("Evens:", evens)

# ── sorted() with key ──
students = [("Alice", 85), ("Bob", 92), ("Charlie", 78)]
by_score = sorted(students, key=lambda s: s[1], reverse=True)
print("By score:", by_score)
````

## 1️⃣2️⃣ List Comprehensions

### 📝 Notes
- Concise way to create lists from sequences.
- Syntax: `[expression for item in iterable if condition]`
- Also: **Dict comprehensions** `{k:v for ...}` and **Set comprehensions** `{expr for ...}`
- **Generator expressions** `(expr for ...)` — memory efficient, lazy evaluation.

````python
# ── List Comprehension ──
# Basic
squares = [x**2 for x in range(1, 6)]
print("Squares:", squares)        # [1, 4, 9, 16, 25]

# With condition
even_sq = [x**2 for x in range(1, 11) if x % 2 == 0]
print("Even squares:", even_sq)   # [4, 16, 36, 64, 100]

# String transformation
words = ["hello", "WORLD", "Python"]
lower = [w.lower() for w in words]
print("Lowercase:", lower)

# Nested comprehension (flattening matrix)
matrix = [[1, 2, 3], [4, 5, 6], [7, 8, 9]]
flat = [num for row in matrix for num in row]
print("Flattened:", flat)

# ── Dict Comprehension ──
word_lengths = {word: len(word) for word in ["apple", "banana", "fig"]}
print("Word lengths:", word_lengths)

# ── Set Comprehension ──
unique_squares = {x**2 for x in [-2, -1, 0, 1, 2]}
print("Unique squares:", unique_squares)

# ── Generator Expression ──
gen = (x**2 for x in range(1, 6))   # lazy, no list created in memory
print("Generator sum:", sum(gen))
````

## 1️⃣3️⃣ Exception Handling

### 📝 Notes
- Errors at runtime are called **Exceptions**.
- Use `try / except / else / finally` to handle them gracefully.
- `else`: runs if `try` succeeded without exception.
- `finally`: runs **always**, regardless of exception.
- You can `raise` your own exceptions.

### Common Exception Types
| Exception | Cause |
|-----------|-------|
| `ZeroDivisionError` | Divide by zero |
| `ValueError` | Wrong value type |
| `TypeError` | Wrong data type |
| `IndexError` | List index out of range |
| `KeyError` | Dictionary key not found |
| `FileNotFoundError` | File doesn't exist |
| `NameError` | Variable not defined |

````python
# ── Basic try / except ──
try:
    result = 10 / 0
except ZeroDivisionError:
    print("Cannot divide by zero!")

# ── Multiple exceptions ──
def safe_convert(value):
    try:
        return int(value)
    except ValueError:
        print(f"'{value}' cannot be converted to int")
        return None
    except TypeError:
        print(f"Type error: expected str, got {type(value).__name__}")
        return None

print(safe_convert("42"))     # 42
print(safe_convert("hello"))  # None
print(safe_convert(None))     # None
````

````python
# ── try / except / else / finally ──
def divide(a, b):
    try:
        result = a / b
    except ZeroDivisionError as e:
        print(f"Error: {e}")
        result = None
    else:
        print(f"Success! {a} / {b} = {result}")  # Only if no exception
    finally:
        print("This always runs (cleanup here)")  # Always runs
    return result

divide(10, 2)
print()
divide(10, 0)

# ── Raising Exceptions ──
def set_age(age):
    if not isinstance(age, int):
        raise TypeError(f"Age must be int, got {type(age).__name__}")
    if age < 0 or age > 150:
        raise ValueError(f"Age {age} is out of valid range (0-150)")
    return age

try:
    set_age("twenty")
except TypeError as e:
    print(f"TypeError: {e}")
````

## 1️⃣4️⃣ File I/O

### 📝 Notes
- Use `open()` to read/write files.
- Always use `with` statement (context manager) — auto-closes file.
- Modes: `'r'` read, `'w'` write (overwrite), `'a'` append, `'x'` create, `'b'` binary.

````python
import os

# ── Writing a File ──
with open("sample.txt", "w") as f:
    f.write("Line 1: Hello, Python!\n")
    f.write("Line 2: File I/O is easy.\n")
    f.writelines(["Line 3: One\n", "Line 4: Two\n"])

print("File written successfully!")

# ── Reading a File ──
with open("sample.txt", "r") as f:
    content = f.read()       # Read entire file as string
print(content)

with open("sample.txt", "r") as f:
    lines = f.readlines()    # Read as list of lines
print(lines)

# ── Reading line by line (memory efficient) ──
with open("sample.txt", "r") as f:
    for line in f:
        print(line.strip())

# ── Appending ──
with open("sample.txt", "a") as f:
    f.write("Line 5: Appended line.\n")

# ── Cleanup ──
os.remove("sample.txt")
print("File removed.")
````

## 1️⃣6️⃣ Object-Oriented Programming (OOP)

### 📝 Notes
- OOP organizes code into **classes** (blueprints) and **objects** (instances).
- **4 Pillars of OOP**: Encapsulation, Inheritance, Polymorphism, Abstraction.
- `__init__` is the **constructor** — runs when object is created.
- `self` refers to the current object instance.
- `__str__` defines how object prints (human-readable).
- `__repr__` defines developer-friendly representation.

````python
# ── Basic Class ──
class Person:
    """Represents a person."""
    
    species = "Homo sapiens"   # Class variable (shared)
    
    def __init__(self, name, age):
        """Initialize Person with name and age."""
        self.name = name       # Instance variable
        self.age = age
    
    def greet(self):
        return f"Hi! I'm {self.name}, {self.age} years old."
    
    def have_birthday(self):
        self.age += 1
        print(f"Happy Birthday {self.name}! Now {self.age}.")
    
    def __str__(self):
        return f"Person(name={self.name}, age={self.age})"
    
    def __repr__(self):
        return f"Person({self.name!r}, {self.age})"


# Creating objects
p1 = Person("Alice", 25)
p2 = Person("Bob", 30)

print(p1.greet())
print(p2.greet())
p1.have_birthday()
print(p1)               # calls __str__
print(repr(p2))          # calls __repr__
print(Person.species)    # Class variable
````

````python
# ── Inheritance ──
class Animal:
    def __init__(self, name, sound):
        self.name = name
        self.sound = sound
    
    def speak(self):
        return f"{self.name} says: {self.sound}!"
    
    def breathe(self):
        return f"{self.name} breathes air."


class Dog(Animal):                    # Inherits from Animal
    def __init__(self, name, breed):
        super().__init__(name, "Woof")  # Call parent's __init__
        self.breed = breed
    
    def fetch(self):
        return f"{self.name} fetches the ball!"


class Cat(Animal):
    def __init__(self, name):
        super().__init__(name, "Meow")
    
    def speak(self):                  # Override parent method
        return f"{self.name} says: Purr... Meow! 🐱"


dog = Dog("Rex", "Labrador")
cat = Cat("Whiskers")

print(dog.speak())      # inherited from Animal
print(dog.breathe())    # inherited from Animal
print(dog.fetch())      # Dog-specific
print(cat.speak())      # Overridden!

# isinstance / issubclass
print(isinstance(dog, Dog))     # True
print(isinstance(dog, Animal))  # True (Dog IS-A Animal)
print(issubclass(Dog, Animal))  # True
````

````python
# ── Encapsulation (Public / Protected / Private) ──
class BankAccount:
    def __init__(self, owner, balance=0):
        self.owner = owner         # Public
        self._bank = "PyBank"      # Protected (convention: don't use outside)
        self.__balance = balance   # Private (name-mangled to _BankAccount__balance)
    
    def deposit(self, amount):
        if amount > 0:
            self.__balance += amount
            print(f"Deposited ₹{amount}. Balance: ₹{self.__balance}")
    
    def withdraw(self, amount):
        if 0 < amount <= self.__balance:
            self.__balance -= amount
            print(f"Withdrew ₹{amount}. Balance: ₹{self.__balance}")
        else:
            print("Insufficient funds!")
    
    @property
    def balance(self):             # Getter
        return self.__balance


acc = BankAccount("Alice", 1000)
acc.deposit(500)
acc.withdraw(200)
print("Balance:", acc.balance)   # Via property
# acc.__balance  → AttributeError (name-mangled)
````

````python
# ── Static & Class Methods ──
class MathUtils:
    pi = 3.14159
    
    @staticmethod
    def add(a, b):                     # No self or cls
        return a + b
    
    @classmethod
    def circle_area(cls, radius):      # cls = the class itself
        return cls.pi * radius ** 2


print(MathUtils.add(5, 3))             # Call without creating object
print(MathUtils.circle_area(7))        # Uses class variable pi
````

````python
# ✅ SOLUTION — Exercise 3

# Part 1: Second largest
def second_largest(lst):
    unique = list(set(lst))
    if len(unique) < 2:
        return None
    unique.sort()
    return unique[-2]

nums = [3, 1, 4, 1, 5, 9, 2, 6, 5, 3]
print(f"List: {nums}")
print(f"Second largest: {second_largest(nums)}")

# Part 2: Remove duplicates preserving order
def remove_duplicates(lst):
    seen = set()
    return [x for x in lst if not (x in seen or seen.add(x))]

original = [1, 2, 2, 3, 4, 3, 5, 1, 6]
print(f"\nOriginal: {original}")
print(f"No duplicates: {remove_duplicates(original)}")

# Part 3: Star pattern
print("\nStar Pattern:")
for i in range(1, 6):
    print("* " * i)
````

### ✏️ Exercise 4: Functions & Logic
1. Write a function to check if a number is prime
2. Write a recursive function to compute Fibonacci numbers
3. Write a function that returns a Caesar cipher of a string

````python
# ✏️ YOUR ATTEMPT — Exercise 4

````

````python
# ✅ SOLUTION — Exercise 4

# Part 1: Prime check
def is_prime(n):
    if n < 2:
        return False
    if n == 2:
        return True
    if n % 2 == 0:
        return False
    for i in range(3, int(n**0.5) + 1, 2):
        if n % i == 0:
            return False
    return True

primes = [x for x in range(2, 50) if is_prime(x)]
print(f"Primes < 50: {primes}")

# Part 2: Fibonacci (recursive + memoized)
from functools import lru_cache

@lru_cache(maxsize=None)
def fib(n):
    if n <= 1:
        return n
    return fib(n-1) + fib(n-2)

fib_sequence = [fib(i) for i in range(15)]
print(f"Fibonacci (first 15): {fib_sequence}")

# Part 3: Caesar Cipher
def caesar_cipher(text, shift):
    result = []
    for ch in text:
        if ch.isalpha():
            base = ord('A') if ch.isupper() else ord('a')
            result.append(chr((ord(ch) - base + shift) % 26 + base))
        else:
            result.append(ch)
    return ''.join(result)

message = "Hello, World!"
encrypted = caesar_cipher(message, 3)
decrypted = caesar_cipher(encrypted, -3)
print(f"\nOriginal:  {message}")
print(f"Encrypted: {encrypted}")
print(f"Decrypted: {decrypted}")
````

### ✏️ Exercise 5: OOP — Build a Library System
Create a simple Library system with:
- A `Book` class with `title`, `author`, `isbn`, `available` attributes
- A `Library` class that can `add_book()`, `borrow_book()`, `return_book()`, and `list_books()`

````python
# ✏️ YOUR ATTEMPT — Exercise 5

````

````python
# ✅ SOLUTION — Exercise 5: Library System

class Book:
    def __init__(self, title, author, isbn):
        self.title = title
        self.author = author
        self.isbn = isbn
        self.available = True
    
    def __str__(self):
        status = "✅ Available" if self.available else "❌ Borrowed"
        return f"  [{self.isbn}] '{self.title}' by {self.author} — {status}"


class Library:
    def __init__(self, name):
        self.name = name
        self.__books = {}           # isbn → Book
    
    def add_book(self, book):
        if book.isbn in self.__books:
            print(f"Book {book.isbn} already exists.")
        else:
            self.__books[book.isbn] = book
            print(f"Added: '{book.title}'")
    
    def borrow_book(self, isbn):
        book = self.__books.get(isbn)
        if not book:
            print(f"Book {isbn} not found.")
        elif not book.available:
            print(f"'{book.title}' is already borrowed.")
        else:
            book.available = False
            print(f"Borrowed: '{book.title}'")
    
    def return_book(self, isbn):
        book = self.__books.get(isbn)
        if not book:
            print(f"Book {isbn} not found.")
        elif book.available:
            print(f"'{book.title}' was not borrowed.")
        else:
            book.available = True
            print(f"Returned: '{book.title}'")
    
    def list_books(self):
        print(f"\n📚 {self.name} — Book Catalog:")
        if not self.__books:
            print("  No books available.")
        for book in self.__books.values():
            print(book)


# ── Demo ──
lib = Library("PyCity Public Library")

lib.add_book(Book("Python Crash Course", "Eric Matthes",   "978-1"))
lib.add_book(Book("Clean Code",          "Robert Martin",  "978-2"))
lib.add_book(Book("The Pragmatic Programmer", "Hunt & Thomas", "978-3"))

lib.list_books()

print()
lib.borrow_book("978-1")
lib.borrow_book("978-1")   # Already borrowed

lib.list_books()

print()
lib.return_book("978-1")
lib.list_books()
````
